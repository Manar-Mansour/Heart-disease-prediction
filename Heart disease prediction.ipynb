# Using SVM

```python
from sklearn import svm
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('heart.data.txt', delimiter = ' ')
print(data)
```

```python
x=data.iloc[:,0:13]
print(x)
```

```python
y=data.iloc[:,13]
print(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
print(len(x_train))
print(len(x_test))
```

```python
model=svm.SVC(kernel='linear')
```

```python
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
print(y_predict)
```

```python
accuracy=model.score(x_test,y_test)
print(accuracy)
```

# Using KNN 

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=180)
```

```python
knn
```

```python
knn.fit(x_train,y_train)
```

```python
knn.predict(x_test)
```

```python
accuracy_knn=knn.score(x_test,y_test)
print(accuracy_knn)
```

# Using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg=LogisticRegression(solver='liblinear',multi_class='ovr')
```

```python
logreg.fit(x_train,y_train)
```

```python
logreg.predict(x_test)
```

```python
accuracy_logreg=logreg.score(x_test,y_test)
print(accuracy_logreg)
```

# Using Decision Trees

```python
from sklearn import tree
```

```python
tr = tree.DecisionTreeClassifier()
```

```python
tr.fit(x_test, y_test)
```

```python
tr.predict(x_test)
```

```python
accuracy_tr=tr.score(x_test,y_test)
print(accuracy_tr)
```

# Using Neural Networks

```python
from sklearn.neural_network import MLPClassifier
```

```python
nn = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(50, 100), random_state=1)
```

```python
nn.fit(x_train,y_train)
```

```python
nn.predict(x_test)
```

```python
accuracy_nn=nn.score(x_test,y_test)
print(accuracy_nn)
```
